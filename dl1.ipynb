Experiment 3: Basic Tensor Operations

```python
#Scalar tensors
a = tf.constant(10)
b = tf.constant(5)

print(a+b)
print(a-b)
print(a*b)
print(a/b)
```

```python
#Vector tensors
a = tf.constant([1, 2, 3])
b = tf.constant([4, 5, 6])
print(a+b)
print(a-b)
print(a*b)
print(a/b)
```

```python
#2-D matrix
m1 = tf.constant([[2, 4],
                  [6, 8]])

m2 = tf.constant([[1, 2],
                  [3, 4]])

print(m1+m2)
print(m1-m2)
print(m1*m2)
print(m1/m2)
```

Experiment 4: Data Types and Shape of Tensors

```python
#scalar
scalar = tf.constant(5, dtype=tf.int32)

print("Scalar Tensor:", scalar)
print("Data Type:", scalar.dtype)
print("Shape:", scalar.shape)
print("Dimensions:", scalar.ndim)

#vector
vector = tf.constant([1, 2, 3, 4], dtype=tf.float32)

print("\nVector Tensor:", vector)
print("Data Type:", vector.dtype)
print("Shape:", vector.shape)
print("Dimensions:", vector.ndim)

#2D MATRIX
matrix = tf.constant([[1, 2, 3],
                      [4, 5, 6]], dtype=tf.float64)

print("\n2D Matrix Tensor:\n", matrix)
print("Data Type:", matrix.dtype)
print("Shape:", matrix.shape)
print("Dimensions:", matrix.ndim)
```

Experiment 5: Converting NumPy Arrays to Tensors

```python
import numpy as np

np_array = np.array([10, 20, 30, 40])
tf_tensor = tf.convert_to_tensor(np_array)

print("NumPy Array:", np_array)
print("TensorFlow Tensor:", tf_tensor)
```
